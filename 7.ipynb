# Шаг 7. Нарисуйте график, который отображает revenue (y) по отношению к UnitPrice (x)

```python
t = online_rt[online_rt['Quantity'] > 0]
online_rt_4 = t.groupby('UnitPrice')['Quantity'].sum()
online_rt_4 = pd.DataFrame(online_rt_4)
online_rt_4['UnitPrice'] = online_rt_4.index
online_rt_4['revenue']= online_rt_4['UnitPrice']*online_rt_4['Quantity']
online_rt_5= online_rt_4[online_rt_4['UnitPrice'] !=0]
sns.jointplot(data=online_rt_5, x="UnitPrice", y="revenue", kind='reg')
```
